```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('recommender_file_met.csv', index_col=0)
df
```

```python
heart_beat = pd.read_csv('data_participant_kg.csv')
heart_beat
```

```python
data = pd.merge(df, heart_beat, left_on='filename', right_on='eid', how='left')
data = data[['eid', 'p21022_x', 'Predictions', 'p102_i0_a0']]
data = data.dropna()
data
```

```python
def get_vo2max(row):
    max = 208 - (0.7 * row['p21022_x'])
    return 15 * (max/row['p102_i0_a0'])

data['vo2max'] = data.apply(get_vo2max, axis=1)
data
```

```python
def get_met_min(row):
    if row['Predictions'] == 1:
        return 0
    elif row['Predictions'] == 0:
        return np.round(row['vo2max']/17.5)

    return False

def get_met_max(row):
    if row['Predictions'] == 1:
        return np.round(row['vo2max']/17.5)
    elif row['Predictions'] == 0:
        return np.round(row['vo2max']/6)

    return False


data['met_min'] = data.apply(get_met_min, axis=1)
data['met_max'] = data.apply(get_met_max, axis=1)
data
```

```python
data.to_csv('met_values.csv')
```
